### âœ… **What is an Abstract Class?**

* A **blueprint class** that cannot create objects directly.
* Contains **one or more abstract methods**.
* Used to enforce that all child classes must implement specific methods.

---

### âœ… **What is an Abstract Method?**

* A method declared in the abstract class but contains **no code**.
* Marked using the `@abstractmethod` decorator.
* Forces child classes to provide their own method implementation.

---

### âœ… **How to Create Abstract Classes in Python?**

```python
from abc import ABC, abstractmethod

class ClassName(ABC):
    @abstractmethod
    def method_name(self):
        pass
```

---

### âœ… **Example Code:**

```python
from abc import ABC, abstractmethod

# Abstract Class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Child Class
class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side * self.side

# Object Creation from Child Class
s = Square(5)
print(s.area())  # Output: 25
```

---

### âŒ **You Cannot Do:**

```python
shape = Shape()  # âŒ Error! Cannot create object from abstract class.
```

---

### âœ… **Why Use Abstract Classes?**

* To **force all child classes** to follow the same method structure.
* Used in **frameworks, libraries, and large projects** for consistency.

---

### âœ… **Real-Life Example (Bank System):**

* Abstract Class: `BankAccount` with abstract methods `deposit()` and `withdraw()`.
* Child Classes: `SavingAccount`, `CurrentAccount` must implement both methods.

# ðŸŽ¯ Why Use `@abstractmethod`?

### Without `@abstractmethod`:
- The method in the parent class is just a **normal method**.
- Child classes can choose to **ignore it**.
- **No rule** forces child classes to implement it.

---

### With `@abstractmethod`:
- Marks the method as a **mandatory rule**.
- Python **forces child classes** to implement the method.
- If a child class doesn't implement it, Python raises an **error** when trying to create an object.

---

## ðŸ“Œ Summary Table:

| Without `@abstractmethod`   | With `@abstractmethod`              |
|------------------------------|-------------------------------------|
| Method is optional for child classes. | Method is mandatory for child classes. |
| No enforcement or rules.     | Strict enforcement using Python itself. |
| Can create object even if method not defined in child class. | Error if method not implemented in child class. |

---
